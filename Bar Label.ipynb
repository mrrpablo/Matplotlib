```python
import matplotlib.pyplot as plt
import numpy as np

Geomatics = ("Pablo", "Sterling", "Nulu")
Gender = {
    "Male": np.array([70, 40, 56]),
    "Female": np.array([16, 23, 40])
}
width = 0.4

fig, ax = plt.subplots()
bottom = np.zeros(3)

for Gen, Gender in Gender.items():
    Plot = ax.bar(Geomatics, Gender, width, label = Gen, bottom = bottom)
    ax.bar_label(Plot, label_type = "edge")
plt.legend()
ax.set_ylabel ("Beginners Value")
ax.set_xlabel ("Beginners Name")
ax.set_title("Number of Gender in Geomatics")
plt.show()
```

```python
State = ["Lagos", "Anambra", "Imo", "Enugu"]
Abuja = [234, 167, 189, 130]
fig, ax = plt.subplots()
bar_con = ax.bar(State, Abuja)
ax.set(ylabel = "Distance Km/hr", xlabel = "States", title = "Distances from Abuja to Different States", ylim = (0, 400))
ax.bar_label(bar_con, fmt = lambda x:f'{x *1.6:.1f} km/hr')
```

```python
np.random.seed(1234567)
Geomatics = ("Pablo", "Sterling", "Nulu", "Collins")
y_pos = np.arange(len(Geomatics))
height = 3 + 20 * np.random.rand(len(Geomatics))
error = np.random.rand(len(Geomatics))

fig, ax = plt.subplots()
hbar = ax.barh(y_pos, height, xerr = error, align = "center")
ax.set_yticks(y_pos, labels = Geomatics)
ax.invert_yaxis()
ax.set_title("Number of Gender in Geomatics")
ax.set_xlabel("Values")
ax.set_ylabel("Names")
ax.bar_label(hbar, fmt = "%.2f")
ax.set_xlim(right = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#variables
np.random.seed(1234567)
LandUseClasses = ["Forest", "Agriculture", "Water", "Barren", "Urban"]
years = [2000, 2010, 2020]
y_pos = np.arange(len(LandUseClasses))

#Generate random counts (heights) for each class in each year
count_2000 = 3 + 20 * np.random.rand(len(LandUseClasses))
count_2010 = 3 + 20 * np.random.rand(len(LandUseClasses))
count_2020 = 3 + 20 * np.random.rand(len(LandUseClasses))

fig, ax = plt.subplots()
# bar width
width = 0.2

ax.barh(y_pos - width, count_2000, height=width, label="2000")
ax.barh(y_pos, count_2010, height=width, label="2010")
ax.barh(y_pos + width, count_2020, height=width, label="2020")

ax.set_yticks(y_pos)
ax.set_yticklabels(LandUseClasses)
ax.invert_yaxis()

ax.set_title("Land Use Change over Time")
ax.set_xlabel("Values")
ax.set_ylabel("Land Use Classes")

ax.legend(title="Year")

ax.set_xlim(right = 30)

plt.show()
```
