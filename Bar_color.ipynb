```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
Beginners = ["Pablo", "Steph", "Collins", "Blake"]
Value = [50, 500, 200, 150]
bar_labels = ["red", "blue", "cyan", "orange"]
bar_colors = ["tab:red", "tab:blue", "tab:cyan", "tab:orange"]
ax.bar(Beginners, Value, label = bar_labels, color = bar_colors) 
ax.set_ylabel ("Beginners Value")
ax.legend(title = "Beginners")
ax.set_title("Beginners Weight")
plt.show()
```
