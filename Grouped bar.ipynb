```python
import matplotlib.pyplot as plt
import numpy as np

Class = ("Built area", "water", "bare soil", "vegetation", "urban")
year = {
    "2000" : (14.98, 11.4, 30.23, 17.4, 25.5),
    "2010" : (26.98, 20.4, 19.23, 21.9, 33.4),
    "2020" : (47.98, 40.4, 40.23, 60.3, 35.1),
} 
x = np.arange(len(Class))
width = 0.3
multiplier = 0

fig, ax = plt.subplots(layout = "constrained")

for attribute, measurement in year.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label= attribute)
    ax.bar_label(rects, padding = 3)
    multiplier += 1

ax.set_ylabel("area in hectares")
ax.set_xlabel("classes of LCLU")
ax.set_title("Land cover Land use Analysis for 2000, 2010, 2020")
ax.set_xticks(x + width, Class)
ax.legend()
ax.set_ylim(0, 70)
ax.grid(color= 'k', linestyle= '-', linewidth=0.1)
plt.show()    
```

# catergorical variable

```python
import matplotlib.pyplot as plt

Class = {"Pablo" : 10, "chika" : 2, "sterling" : 34, "erus" : 46}
name = list(Class.keys())
networth = list(Class.values())

fig, ax = plt.subplots(1,3, figsize = (9,3), sharey= True)
ax[0].bar(name, networth)
ax[1].scatter(name, networth)
ax[2].plot(name, networth)
```

# pie chart

```python
Class = ("Built area", "water", "bare soil", "vegetation", "urban")
year_2000 = (14.98, 11.4, 30.23, 17.4, 25.5)
fig, ax = plt.subplots()
ax.pie(year_2000, labels = Class)
```

```python
fig, ax = plt.subplots()
ax.pie(year_2000, labels = Class, autopct = "%1.1f")
```

# color coding

```python
fig, ax = plt.subplots()
ax.pie(year_2000, labels = Class, colors=(["r", "b", "y", "k", "g"]))
```

# hatching

```python
fig, ax = plt.subplots()
ax.pie(year_2000, labels = Class, hatch=(["**0", "o0", "o.o", ".^.", ".||."]))
```
